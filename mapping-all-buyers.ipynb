```python
# Make an empty map
desti_m = folium.Map(location=[df.Lat.min(), df.Lon.max()], tiles="OpenStreetMap", zoom_start=12)

# I can add marker one by one on the map
for i in range(df.destination_pincode.nunique()):
    folium.Marker([df.iloc[i]['Lat'], df.iloc[i]['Lon']],popup=df.iloc[i]['destination_pincode'],icon=folium.Icon(color="green")).add_to(desti_m)
#m.save('Folium BBMP.png')  #<-----To download the graph
desti_m
```

```python
cat = pd.read_csv("../input/destination-and-catergory/Serviced pincodes of all categories - Sheet1 (1).csv")
cat.head()
```

We can select any particular product category and see location of completed orders in that product category 

```python
#specify category to see destination location
#here we specify which category dispatched location we want to see
category = "COMPOSITE PAPER AND PAPERBOARD (MADE BY STICKING FLAT LAYERS OF PAPER OR PAPERBOARD TOGETHER WITH AN ADHESIVE), NOT SURFACECOATED OR IMPREGNATED, WHETHER OR NOT INTERNALLY REINFORCED, IN ROLLS OR SHEETS"
```

```python
temp_cat = cat.loc[(cat['Produt category'] == category)]
temp_cat.shape[0]
```

```python
# Make an empty map
cat_m = folium.Map(location=[temp_cat.lat.min(), temp_cat.lon.max()], tiles="OpenStreetMap", zoom_start=12)

# I can add marker one by one on the map
for i in range(temp_cat.destination_pincode.nunique()):
    folium.Marker([temp_cat.iloc[i]['lat'], temp_cat.iloc[i]['lon']], popup=temp_cat.iloc[i]['destination_pincode'],icon=folium.Icon(color="green")).add_to(cat_m)
#m.save('Folium BBMP.png')  #<-----To download the graph
cat_m
```
