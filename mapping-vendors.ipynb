```python
# Make an empty map
origin_m = folium.Map(location=[df.lat.min(), df.lon.max()], tiles="OpenStreetMap", zoom_start=12)

# I can add marker one by one on the map
for i in range(df.origin_pincode.nunique()):
    folium.Marker([df.iloc[i]['lat'], df.iloc[i]['lon']], popup=df.iloc[i]['Unique vendors']).add_to(origin_m)
#m.save('Folium BBMP.png')  #<-----To download the graph
origin_m
```
